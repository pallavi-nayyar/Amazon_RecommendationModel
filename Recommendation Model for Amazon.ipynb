# Divide the data into training and test data

```python
!pip install scikit-surprise
```

```python
from surprise import Reader, Dataset
```

```python
reader = Reader(line_format='user item rating', rating_scale=(0.5,5))
```

```python
reader
```

```python
meltdf.columns
```

```python
data = Dataset.load_from_df(meltdf[['user_id', 'Movies', 'ratings']], reader=reader)
```

```python
data
```

```python
data.df.head()
```

```python
from surprise import KNNBasic
```

```python
knn_model = KNNBasic(k=15, min_k=5, sim_options = {'name': 'cosine', 'user_based': True}, verbose=True)
```

```python
from surprise.model_selection import cross_validate
```

```python
results = cross_validate(algo=knn_model, data=data, measures= ['rmse','MAE'], cv=3, return_train_measures=True)
```

```python
print(results['test_rmse'].mean())
```

```python
user_id="A1CV1WROP5KTTW"
Movies = 4
ratings='5'
val = knn_model.predict(uid=user_id, iid=Movies,r_ui=ratings)
```

```python
print(f'Prediction of rating for user {userid} and movie {Movies} is {np.round(val,0)}')
```
